```python
import pandas as pd
import numpy as np
from data_pre_proc import read_data, filter_data, filter_by_earnigs
```

```python
df = read_data('https://www.dropbox.com/s/3avq0b4fsbujtvp/pequiv_long.dta?dl=1')
```

```python
data_dict = {
        # Identifiers
        'year' : "Year",
        'x11101LL' : 'id',
        'x11102' : 'Household id',

        # Demographic variables
        "d11101" : "Age",
        "d11102LL" : "Sex",
        "h11112" : "Spouse in HH",
        "d11108" : "Education",
        "d11109" : "Race",
        "d11106" : "Household Size",

        # Employment variables
        "e11101" : "Annual Work Hours",
        "e11102" : "Employment Status",
        "e11105" : "Ocupation",

        # Income variables
        "i11113" : "After-tax Income"
        }
```

```python
df_1 = filter_data(df, 1978, 1997, data_dict)
```

```python
df_2 = filter_by_earnigs(df_1, "After-tax Income", 5000, 5)
df_2["Spouse in HH"] = df_2["Spouse in HH"].apply(lambda x: 1 if x == df_2["Spouse in HH"].values[0] else 0)
df_2['HS dropout'] = df_2['Education'].apply(lambda x: 1 if x == "Less than H S  1" else 0)
df_2['HS graduate'] = df_2['Education'].apply(lambda x: 1 if x == "High School    2" else 0)
# df_2['College dropout'] = df_2['Education'].apply(lambda x: 1 if x == "More than H S  3" else 0)
```
